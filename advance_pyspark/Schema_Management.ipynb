```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Spark SQL').getOrCreate()
sc = spark.sparkContext
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
data1= [ ("James","","Smith", "36636", "M", 3000),
          ("Michael", "Rose","", "40288", "M", 4000),
          ("Robert","", "Williams", "42114", "M", 4000),
          ("Maria","Anne", "Jones", "39192", "F", 4000),
          ("Jen","Mary", "Brown", "37371", "F", -1)
]
```

```python
# firstname cant accept null value
schema1= StructType([\
                      StructField("firstname", StringType(),False),\
                      StructField("middlename", StringType(),True),\
                      StructField("lastname", StringType(),True),\
                      StructField("id", StringType(),True),\
                      StructField("gender", StringType(),True),\
                      StructField("salary", StringType(),True),\
                     ])
```

```python
df = spark.createDataFrame(data=data1, schema=schema1)
```

```python
df.show()
```

# Nested Schema Definiton

```python
data2= [ (("James","","Smith"), "36636", "M", 3000),
          (("Michael", "Rose",""), "40288", "M", 4000),
          (("Robert","", "Williams"), "42114", "M", 4000),
          (("Maria","Anne", "Jones"), "39192", "F", 4000),
          (("Jen","Mary", "Brown"), "37371", "F", -1)
          ]
```

```python
schema2= StructType([
                      StructField("name", StructType([
                                                StructField("firstname", StringType(),False),
                                                StructField("middlename", StringType(),True),
                                                StructField("lastname", StringType(),True),
                                  ])),
                      StructField("id", StringType(),True),
                      StructField("gender", StringType(),True),
                      StructField("salary", StringType(),True),
])
```

```python
df2 = spark.createDataFrame(data=data2, schema=schema2)
df2.show()
```

# Array Type and Map Type

```python
from pyspark.sql.types import *
```

```python
array_data= [ (("James","","Smith"),["music", "playing" ,"chess"], { 'salary': '3000', 'dept' : 'HR'}),
          (("Michael", "Rose",""), ["music", "reading" ,"chess"], { 'salary': '3000', 'dept' : 'HR'}),
          (("Robert","", "Williams"), ["music", "playing" ,"chess"], { 'salary': '3000', 'dept' : 'HR'}),
          (("Maria","Anne", "Jones"),["music", "playing" ,"chess"], { 'salary': '3000', 'dept' : 'HR'}),
          (("Jen","Mary", "Brown"), ["music", "playing" ,"chess"], { 'salary': '3000', 'dept' : 'HR'})
          ]
```

```python
array_schema= StructType([
                      StructField("name", StructType([
                                                StructField("firstname", StringType(),False),
                                                StructField("middlename", StringType(),True),
                                                StructField("lastname", StringType(),True),
                                  ])),

                      StructField("hobbies", ArrayType(StringType()),True),
                      StructField("properties", MapType(StringType(),StringType()),True),
])
```

```python
array_df= spark.createDataFrame(data=array_data, schema=array_schema)
array_df.show()
```
